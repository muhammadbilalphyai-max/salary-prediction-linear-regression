```python
#Predication of Salaries Based on Years of  Experience

#Import important Libraries
import numpy as np
import pandas as pd
from sklearn import linear_model
from matplotlib import pyplot as plt


# Load data set

df=pd.read_csv(r"C:\Users\M BILAL\Desktop\Salary_dataset.csv")

#Creation of object from class
lr=linear_model.LinearRegression()
# Model_Training
lr.fit(df[["YearsExperience"]],df.Salary)

#Regression Line
plt.title('Experience vs Salaries')
plt.xlabel("Years _of_Experience")
plt.ylabel("Salary")
plt.scatter(df[["YearsExperience"]],df.Salary)
plt.plot(df[["YearsExperience"]],lr.predict(df[["YearsExperience"]]),color="red")



```

```python
#Testing
experience=float(input('Enter Experience in Years'))
predicted_salary=lr.predict([[experience]])
print(f"The predicted salary for {experience} years of experience is Rs {predicted_salary}")
model_accurracy=lr.score(df[["YearsExperience"]],df.Salary)
print(f"Model Accuracy :{model_accurracy*100} %")
```
